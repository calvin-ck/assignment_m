#### 4) insert / update / delete

```python
# insert brand Y
i = """{"action" : "insert","data" : {"브랜드":"Y", "상의":1000, "아우터": 5000, "바지":4000, "스니커즈":9900, "가방":2200, "모자":1000, "양말":1000, "액세서리":2000}}"""
res4, msg = fr.upsertData(i)
if res4 == False:
    print (msg)
fr.fashion_list
```

```python
# update brand A
u = """{"action" : "update","data" : {"브랜드":"A", "모자":12000, "액세서리":2000}}"""
res4, msg = fr.upsertData(u)
if res4 == False:
    print (msg)
fr.fashion_list
```

```python
# delete brand B
d = """{"action" : "delete","data" : {"브랜드":"B"}}"""
res4, msg = fr.upsertData(d)
if res4 == False:
    print (msg)
fr.fashion_list
```

```python
# error case 1 : 내용 없음 
e1 = """    """
res4, msg = fr.upsertData(e1)
if res4 == False:
    print (msg)
```

```python
# error case 2 : 삭제 브랜드 없음 
e2 = """{"action" : "delete","data" : {"브랜드":"X"}}"""
res4, msg = fr.upsertData(e2)
if res4 == False:
    print (msg)
```

```python
# error case 3 : jsen error 
e3 = """{"action" : "delete","data" : {"브랜드":"""
res4, msg = fr.upsertData(e3)
if res4 == False:
    print (msg)
```
